```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a DataFrame

data = pd.read_csv('D:\\software\\python\\Datasets\\titanic_train.csv')
```

```python
# Check for missing values
missing_values = data.isna().sum()

# Drop rows with missing values
data_cleaned = data.dropna()

# Check if missing values have been removed

print("Number of missing values after cleaning:")
print(data_cleaned.isna().sum())
```

```python
# Create a box plot for the 'Age' column

sns.boxplot(x='Age', data=data_cleaned)
plt.show()

# Define a function to remove outliers based on a threshold

def remove_outliers(data, column, threshold):
    return data[(data[column] < threshold[0]) | (data[column] > threshold[1])]

# Define an age threshold based on the box plot
age_threshold = (0, 80)

# Remove outliers from the 'Age' column
data_cleaned = remove_outliers(data_cleaned, 'Age', age_threshold)

# Check if outliers have been removed
sns.boxplot(x='Age', data=data_cleaned)
plt.show()
```
